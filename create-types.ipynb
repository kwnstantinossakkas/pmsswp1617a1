# Notebook για την παραγωγή των δεδομένων N-Triples (1η εργασία).

Χρησιμοποιήστε τον κώδικα από το [notebook του 5ου εργαστηρίου](https://github.com/mixstef/courses/blob/master/pms-swp/schedule4.ipynb) με τις κατάλληλες αλλαγές για να φτιάξετε το αρχείο `schedule-data.nt` της **1ης εργασίας**.

Προσθέστε τον κώδικά σας στα επόμενα κελιά.

```python
import csv

rdf = "http://www.w3.org/1999/02/22-rdf-syntax-ns#"
voc = "http://dilab77.ionio.gr/swp/sakkas/vocab#"
rdftype = "<" + rdf + "type" + ">"

triples = []
with open('schedule-uris.csv','r',newline='') as ifp:
    reader = csv.reader(ifp)
    for row in reader:
        triples.append(row)
        #print(row)
#print(triples)

# ΠΡΟΣΟΧΗ! ΠΛΗΡΗΣ ΛΥΣΗ
# fix objects (literals and uris)
visited = set()
triples2 = []
index = 0
for s,p,o in triples:        
    #fix subjects
    s = s[0:2] + "bnode" +  s[2]
    # add RDF declarations for subjects, once !
    if s not in visited:        
        voddial = "<" + voc + "Διάλεξη" + ">"
        triples2.append('{} {} {} .'.format(s,rdftype,voddial))
        visited.add(s)
    
    # fix URI predicates
    p = '<'+ p + '>'

    # εάν είναι ώρα, νέο ο είναι στη μορφή "15:00:00"^^<http://www.w3.org/2001/XMLSchema#time>
    # αν είναι άλλο literal, νέο o είναι στη μορφή "κείμενο"
    # αλλιώς νέο ο είναι URI στη μορφή <http://..κλπ>

    # fix all objects    
    if o.endswith(":00"): # είναι ώρα
        o1 = o[2:] + ":00"
        o2 = "^^<http://www.w3.org/2001/XMLSchema#time>"
        o = '"{}"{}'.format(o1,o2)    	  
    elif o.startswith("http:"): # είναι URI object
        o = "<" + o + ">"
        if o not in visited:
            if index % 7 == 3:
                vocdidask = "<" + voc + "Διδάσκων" + ">"
                triples2.append('{} {} {} .'.format(o,rdftype,vocdidask))
                visited.add(o)
            elif index % 7 == 4:
                vocaithusa = "<" + voc + "Αίθουσα" + ">"
                triples2.append('{} {} {} .'.format(o,rdftype,vocaithusa))
                visited.add(o)
            elif index % 7 == 6:
                vocmathima = "<" + voc + "Μάθημα" + ">"
                triples2.append('{} {} {} .'.format(o,rdftype,vocmathima))
                visited.add(o)
    else: # είναι το αναγνωριστικό του εξαμήνου
        o = '"{}"'.format(o[2:])

    # add triple to new list
    triples2.append('{} {} {} .'.format(s,p,o))
    index += 1
#print(triples2)

# Προσθήκη RDF λέξεων

# αποθηκεύστε την τελική λίστα εδώ
with open('schedule-data.nt','w',newline='') as ofp:
    for row in triples2:
        ofp.write(row + '\n')
        print(row)
```
